```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import matplotlib.cm as cm
from sklearn.feature_selection import f_regression
from statsmodels.stats.anova import anova_lm
```

```python
data = pd.read_csv('../data/NBA_Players.csv')
data
```

#### Salary v.s. (Stats & Titles)

```python
features_stats = data[['G', 'GS', 'FG_Prct', 'Three_P_Prct', 'Two_P_Prct',
                       'eFG_Prct', 'FT_Prct', 'TRB', 'AST', 'STL',
                       'BLK', 'TOV', 'PF']]
features_title = data[['Potw', 'APG_Leader', 'MVP', 'PPG_Leader',
                       'RPG_Leader', 'Rookie', 'WS_Leader']]
salary = data['Salary']
features_title.head()
```

```python
#scatter plot
plt.figure(figsize = (18, 9))

# i: index
for i,  col in enumerate(features_stats.columns):
    plt.subplot(3, len(features_stats.columns) / 2, i + 1)
    plt.subplots_adjust(hspace = 0.5)
    x = features_stats[col]
    y = salary
    plt.plot(x, y, 'o')
    # Create regression line
    plt.plot(np.unique(x),  np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Salary')
```

```python
#scatter plot
plt.figure(figsize = (12, 6))

# i: index
for i,  col in enumerate(features_title.columns):
    plt.subplot(2, len(features_title.columns) / 2 + 1, i + 1)
    plt.subplots_adjust(hspace = 0.5)
    x = features_title[col]
    y = salary
    plt.plot(x, y, 'o')
    # Create regression line
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Salary')
```

```python
#correlation
columns = ['Salary', 'G', 'GS', 'FG_Prct', 'Three_P_Prct', 
           'Two_P_Prct', 'eFG_Prct', 'FT_Prct', 'TRB', 'AST', 
           'STL', 'BLK','TOV', 'PF', 'Potw', 
           'APG_Leader', 'MVP', 'PPG_Leader',
           'RPG_Leader', 'Rookie',  'WS_Leader']
all_data = data[columns]
import seaborn as sns
correlation_matrix = all_data.corr().round(2)
# annot = True to print the values inside the square
plt.figure(figsize = (15, 10))
ax = sns.heatmap(data = correlation_matrix, annot = True, vmin = -1, vmax = 1, center = 0)
ax.set_ylim(len(columns) + 0.5, - 0.5)
plt.show()
```

```python
formula_1 = f'Salary ~ {" + ".join(list(features_stats.columns))}'
lm1 = smf.ols(formula_1, data = data).fit()
lm1.summary()
```

```python
formula_2 = f'Salary ~ {" + ".join(list(features_title.columns))}'
lm2 = smf.ols(formula_2, data = data).fit()
lm2.summary()
```
